```python
no_params = {'consumer_key': '', 'access_token': '', 'tag': 'n'}
no_result = requests.post('https://getpocket.com/v3/get', data=no_params)
```

```python
no_result.text
```

```python
no_jf = json.loads(no_result.text)
```

```python
no_jf.keys()
```

```python
no_jd = no_jf['list']
no_jd.values()
```

```python
no_urls = []
for num in no_jd.values():
    no_urls.append(num.get('resolved_url'))
no_urls
```

```python
no_df = pd.DataFrame(no_urls, columns = ['urls'])
```

```python
no_df = no_df.assign(wanted = lambda x: 'n')
```

```python
no_df
```

```python
y_params = {'consumer_key': '', 'access_token': '', 'tag': 'y'}
y_result = requests.post('https://getpocket.com/v3/get', data=no_params)
y_jf = json.loads(y_result.text)
y_jd = y_jf['list']
y_urls = []
for num in y_jd.values():
    y_urls.append(num.get('resolved_url'))
y_urls
```

```python
y_df = pd.DataFrame(y_urls, columns = ['urls'])
y_df = y_df.assign(wanted = lambda x: 'y')
y_df
```

```python
df = pd.concat([y_df, no_df], ignore_index=True)
```
